Define Class Game

```python
class Game():
    totalscore=0
    over=False
    frames=[]  
    def __init__(self, name):
        self.name=name       
    def addroll(self, pins):
        if self.over==True:
            print('Game is over!')
            print('Your score is: '+str(self.totalscore))
            return
        
#First roll of a game
        if len(self.frames)==0:
            self.frames.append(Frame())
            self.frames[-1].lst=[]
            self.frames[-1].score=0
            self.frames[-1].lst.append(pins)
            self.frames[-1].score+=pins
            
#Frame 10, second roll
        elif len(self.frames)==10 and len(self.frames[-1].lst)==1:
            self.frames[-1].lst.append(pins)
            self.frames[-1].score+=pins
#frame 10, third roll
        elif len(self.frames)==10 and len(self.frames[-1].lst)==2:
            if sum(self.frames[-1].lst)==10 or self.frames[-1].lst[-1]==10:
                self.frames[-1].lst.append(pins)
                self.frames[-1].score+=pins
                self.over=True
            else:
                self.over=True
                return
            
#no strike nor two rolls in last frame
        elif self.frames[-1].lst[-1]!=10 and len(self.frames[-1].lst)<2:
            self.frames[-1].lst.append(pins)
            self.frames[-1].score+=pins
            try:
                self.frames[-2]
            except:
                pass
            else:
                if self.frames[-2].lst[-1]==10:
                    self.frames[-2].score+=pins                
            
#strike or two rolls in last frame
        elif self.frames[-1].lst[-1]==10 or len(self.frames[-1].lst)==2:
            self.frames.append(Frame())
            self.frames[-1].lst=[]
            self.frames[-1].score=0
            self.frames[-1].lst.append(pins)
            self.frames[-1].score+=pins
#strike last roll
            if self.frames[-2].lst[-1]==10:
                self.frames[-2].score+=pins
#strike last two rolls
                try:
                    self.frames[-3]
                except:
                    pass
                else:
                    if self.frames[-3].lst[-1]==10:
                        self.frames[-3].score+=pins
#spare
            elif sum(self.frames[-2].lst)==10 and len(self.frames[-1].lst)==1:
                self.frames[-2].score+=pins

            
        self.totalscore=0
        print('Number of Frames: ' + str(len(self.frames)))
        for i in self.frames:
            self.totalscore+=i.score
            print('('+str(i.lst) + str(i.score)+')')
```
